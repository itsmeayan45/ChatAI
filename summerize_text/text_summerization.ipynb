## prompt template text summerization

```python
from langchain_classic.chains import LLMChain
from langchain_classic import PromptTemplate
genereictemplate="""
    write a summary of the following speech:
    speech:{speech}
    Translate the precise summery to {language}
"""
prompt=PromptTemplate(
    input_variables=['speech','language'],
    template=genereictemplate
)
prompt
```

```python
complete_prompt=prompt.format(speech=speech,language="French")
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain=LLMChain(llm=llm, prompt=prompt)
summary=llm_chain.run({'speech':speech,'language':"bengali"})
summary
```
